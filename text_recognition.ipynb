```python
# import the necessary packages
from PIL import Image
import pytesseract
import cv2
import os
 
settings={}
settings['preprocess']="blur"

for i in range(1,8):
    
    img = cv2.imread('./samples/bib/bib'+str(i)+'.png',1)
    plt.imshow(img)
    plt.show()
    # load the example image and convert it to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # check to see if we should apply thresholding to preprocess the image
    if settings['preprocess'] == "thresh":
        gray = cv2.threshold(gray, 0, 255,
            cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

    # noise
    elif settings['preprocess'] == "blur":
        gray = cv2.medianBlur(gray, 3)

    filename = "gray"+str(i)+".png"
    cv2.imwrite(filename, gray)
    
    text = pytesseract.image_to_string(Image.open('./samples/bib/bib'+str(i)+'.png'))
    os.remove(filename)
    
    print(text)

    # show the output images
    #cv2.imshow("Image", img)
    #cv2.imshow("Output", gray)
    #cv2.waitKey(0)
```
